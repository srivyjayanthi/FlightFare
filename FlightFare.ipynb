```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#1. Read data
df = pd.read_csv(r'C:\Users\vyjay\Documents\Flight_Fare\Clean_Dataset.csv')
```

```python
# Prints number of rows and columns
print(df.shape)
```

```python
# Prints the column names
print(df.columns)
```

```python
# Print first five rows
print(df.head())
```

```python
#2. Get data description
print(df.describe())
```

```python
#3. Get duplicate count
print(df.duplicated().sum()) 
```

```python
#4. Get missing value count for each column
print(df.isnull().sum())
```

```python
#5. Plot graphs for each column 
for col in df.columns:
    if df[col].dtype == 'object': 
        plt.figure()
        df[col].value_counts().plot.pie(autopct='%1.1f%%')
        plt.title(col)
    else:
        plt.figure()
        df[col].plot.hist()
        plt.title(col)
plt.show()
```
